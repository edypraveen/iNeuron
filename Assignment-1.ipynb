### 2. Create the below pattern using nested for loop in Python

```python
n = 5

for i in range(0,n):
    for j in range (0,i+1):
        print('* ',end="")
    print('\n')
    
for k in range(n,0,-1):
    for l in range(k-1,0,-1):
        print('* ', end="")
    print('\n')
```
